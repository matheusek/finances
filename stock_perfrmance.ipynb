```python
import pandas as pd
import numpy as np
import yfinance as yf
from sklearn.linear_model import LinearRegression
from datetime import datetime, timedelta
import plotly.express as px

# Definindo as ações a serem analisadas
acoes = ['AAPL', 'GOOG', 'TSLA']

# Definindo o período de tempo a ser analisado
start_date = datetime.today() - timedelta(days=365)
end_date = datetime.today()

# Obtendo os dados históricos dos preços das ações
precos_acoes = yf.download(acoes, start_date, end_date)['Close']

# Criando uma lista vazia para armazenar os resultados
resultado = []

# Analisando cada ação e calculando a variação
for acao in acoes:
    # Calculando o preço médio da ação
    preco_medio = precos_acoes[acao].mean()

    # Obtendo o preço atual da ação
    preco_atual = precos_acoes[acao][-1]

    # Calculando a variação da ação
    variacao = (preco_atual - preco_medio) / preco_medio

    # Armazenando os resultados em uma lista
    resultado.append({'Ação': acao, 'Preço Atual': preco_atual, 'Preço Médio': preco_medio, 'Variação': variacao})

# Transformando a lista em um DataFrame
resultado_df = pd.DataFrame(resultado, columns=['Ação', 'Preço Atual', 'Preço Médio', 'Variação'])

# Criando um gráfico de linhas interativo com as variações das ações
fig = px.line()

for acao in acoes:
    fig.add_scatter(x=precos_acoes[acao].index, y=precos_acoes[acao].values, name=acao)

# Adicionando um título e legendas aos eixos
fig.update_layout(title='Variação das Ações', xaxis_title='Data', yaxis_title='Preço')

# Exibindo o gráfico
fig.show()
```
